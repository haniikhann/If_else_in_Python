```python
a = 2456789
num = a
b = 0
d = 0
for i in str(a):
    c = int(i)%10
    if c%2 == 0:
        b+=1
    else:
        d+=1

print(f'Even numbers in {num}: {b}')
print(f'Odd numbers in {num}: {d}')
```

```python
even_sum = 0
odd_sum = 0
for i in range(1,11):
    if i%2 == 0:
        even_sum+=i
    else:
        odd_sum+=i
    
print(f'Sum of even numbers (1-10): {even_sum}')
print(f'Sum of odd numbers (1-10): {odd_sum}')
```

```python
a = 102030
b = 0
for i in str(a):
    if int(i)%10==0 :
        continue
    else:
        c = int(i)%10
        b = b*10+c
    
print(f'{a} without 0: {b}')
```
